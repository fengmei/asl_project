```python
import os
import tarfile
import re
import pprint


### prior to running this script run below to get the enron email dataset 
"""
wget https://www.cs.cmu.edu/~./enron/enron_mail_20150507.tar.gz
"""

# Path to the Enron email dataset - assume it's in the same path as this script
dataset_path = "enron_mail_20150507.tar.gz"

script_dir = os.path.dirname("./")
extracted_dir = os.path.join(script_dir, "extracted_dataset")
body_pattern = r"(?<=\n\n)([\s\S]*?)(?=\n\n|$)(?!.*?-----Original Message-----)" # we want to ignore the Original Message that will go on it on it on 
from_to_pattern = r"From: ([^\n]+)\nTo: ([^\n]+)"
subject_pattern = r"Subject: ([^\n]+)"

# Extract the dataset if it hasn't been extracted before
if not os.path.isdir(extracted_dir):
    with tarfile.open(dataset_path, "r:gz") as tar:
        tar.extractall(extracted_dir)

def extract_email_info(file_path):
    with open(file_path, "r", encoding="latin-1") as f:
        email_content = f.read()
        bodies = re.findall(body_pattern, email_content)
        from_to_matches = re.findall(from_to_pattern, email_content)
        subjects = re.findall(subject_pattern, email_content)

        if bodies and from_to_matches and subjects:
            from_email, to_email = from_to_matches[0]
            return bodies, from_email, to_email, subjects[0]

    return None, None, None, None

body_list = []
should_exit = False
for root, dirs, files in os.walk(extracted_dir):
    i = 0
    for file in files:
        i += 1
        file_path = os.path.join(root, file)
        bodies, from_email, to_email, subject = extract_email_info(file_path)
        if bodies and from_email and to_email and subject:
            # don't print too much
            if i >= 1000:
                should_exit = True
                break
            result_dict = {}
            whole_body = ""
            for body in bodies:
                whole_body = "".join([whole_body, body])
                # print("Body:")
                # print(body)
                # print("-" * 50)
            clean = whole_body
            clean = clean.split('-----Original Message-----')
            clean = clean[0]

            result_dict["body"] = clean
            result_dict["from"] = from_email
            result_dict["to"] = to_email
            result_dict["subject"] = subject
            
            body_list.append(result_dict)
            # print("From:", from_email)
            # print("To:", to_email)
            # print("Subject:", subject)
            # print("=" * 50)
    if should_exit:
        break

pp = pprint.PrettyPrinter(indent=4)
pp.pprint(body_list[:10])

```
